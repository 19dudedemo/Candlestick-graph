```python
import pandas_datareader as data
from bokeh.plotting import figure,output_file,show

import datetime
start=datetime.datetime(2019,1,1)
end=datetime.datetime(2019,1,10)

df=data.DataReader(name="GOOG",data_source="yahoo",start=start,end=end)
```

```python
df
```

```python
def inc_dec(c,o):
    if c>o:
        value="Increase"
    elif c<o:
        value="Decrease"
    else:
        value="Equal"
    return value

df["Status"]=[inc_dec(c,o) for c,o in zip(df.Close,df.Open)]
```

```python
df["Middle"]=(df.Close+df.Open)/2
```

```python
df["Hight"]=abs(df.Close-df.Open)

df
```

```python


p=figure(x_axis_type='datetime',width=1000,height=500)

p.title.text="Finace Chart"




hours_12=12*60*60*1000

p.segment(df.index,df.Open,df.index,df.Close,line_color="black")

p.rect(df.index[df.Status=="Increase"],df.Middle[df.Status=="Increase"],
       hours_12,df.Hight[df.Status=="Increase"],fill_color="green",line_color="black")

p.rect(df.index[df.Status=="Decrease"],df.Middle[df.Status=="Decrease"],
       hours_12,df.Hight[df.Status=="Decrease"],fill_color="red ",line_color="black")
```

```python
output_file("CS.html")
```
